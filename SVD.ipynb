```python
import numpy as np

# Sample matrix (3x2)
A = np.array([[1, 2],
              [3, 4],
              [5, 6]])
```

```python
# Perform SVD
U, S, VT = np.linalg.svd(A)

print("U:\n", U)
print("\nS (singular values):\n", S)
print("\nVT (V transposed):\n", VT)
```

```python
# Convert S to diagonal matrix
S_matrix = np.zeros((U.shape[1], VT.shape[0]))
np.fill_diagonal(S_matrix, S)
```

```python
# Multiply U × S × VT to get original matrix
A_reconstructed = np.dot(U, np.dot(S_matrix, VT))
print("\nReconstructed A:\n", A_reconstructed)
```
