```python
#Example: Classify flowers (Iris dataset again, 2 features)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, datasets
from sklearn.model_selection import train_test_split
from matplotlib.colors import ListedColormap
```

```python
# Load the iris dataset and use only first 2 features
iris = datasets.load_iris()
X = iris.data[:, :2]
y = iris.target
```

```python
# Only use 2 classes for clear separation (Setosa & Versicolor)
X = X[y != 2]
y = y[y != 2]
```

```python
# Split into training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Create SVM model with linear kernel
model = svm.SVC(kernel='linear')
model.fit(X_train, y_train)
```

```python
# Create mesh for decision boundary
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500),
                     np.linspace(y_min, y_max, 500))

Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
```

```python
# Plotting
plt.figure(figsize=(10, 6))
cmap_light = ListedColormap(['#FFAAAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#0000FF'])

plt.contourf(xx, yy, Z, cmap=cmap_light, alpha=0.4)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold, edgecolors='k', s=70)
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title("SVM Decision Boundary (Linear Kernel)")
plt.grid(True)
plt.show()
```

```python
What You’ll See:
🟦🟥 Color regions: area belonging to each class

🌼 Dots: real flower data

✂️ SVM tries to slice the data cleanly between the two classes

```
